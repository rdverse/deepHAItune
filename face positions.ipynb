```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from random import random as rand
```

```python
for activity in ['sleeping', 'standing', 'sitting']:
    
    df = pd.read_csv('/home/drdev/Documents/face_touch_clean/face_positions/standing.csv'.format(acitivity))

    plt.figure(figsize = (20,5))
    ax = plt.subplot()
    ax.tick_params(axis = 'both', direction = 'inout')

    plt.title('Touching face from standing position', fontsize = 16)
    ax.plot(df.Accel_WR_X_CAL,label= 'x')
    ax.plot(df.Accel_WR_Y_CAL,label= 'y')
    ax.plot(df.Accel_WR_Z_CAL,label= 'z')
    plt.xlabel('Time', fontsize = 14)
    plt.ylabel('Acceleration', fontsize = 14)

    plt.legend()
    plt.grid()
    plt.savefig('../plots/face_touch_positions/sleeping.jpg')
    plt.show()
    %matplotlib inline
```
