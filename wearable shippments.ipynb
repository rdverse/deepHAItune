```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize = (10,10))
plt = plt.subplot(111)
barWidth = 0.25

smartwatch = [69.3, 91.4, 156]
wristband = [68.2, 67.7, 74.4]
others = [28.2, 2.6,4.4]

indices = np.arange(len(smartwatch)+1)

smW = [0.25, 0.5, 0.75]
wbW = [1.25, 1.5, 1.75]
oW = [2.25, 2.5, 2.75]

# plt.bar(smW, smartwatch, width = barWidth, color = '#2c7bb6', hatch = '/', edgecolor= 'black')
# plt.bar(wbW, wristband, width = barWidth, color = '#abd9e9', hatch = 'o', edgecolor= 'black')
# plt.bar(oW, others, width = barWidth, color = '#fc8d59', hatch = '*', edgecolor= 'black')

feats = {
    'color' : ['#2c7bb6','#abd9e9','#fc8d59'],
    'hatch' : ['/', 'o','*']
}

for i in range(3):
    if i ==0:
        plt.bar(smW[i], smartwatch[i], width = barWidth, color = feats['color'][i], hatch = feats['hatch'][i], edgecolor= 'black', label = '2019')
        plt.bar(wbW[i], wristband[i], width = barWidth, color = feats['color'][i], hatch =feats['hatch'][i], edgecolor= 'black', label = None)
        plt.bar(oW[i], others[i], width = barWidth, color = feats['color'][i],hatch =feats['hatch'][i], edgecolor= 'black', label = None )
    
    elif i==1:
        plt.bar(smW[i], smartwatch[i], width = barWidth, color = feats['color'][i], hatch = feats['hatch'][i], edgecolor= 'black', label ='2020')
        plt.bar(wbW[i], wristband[i], width = barWidth, color = feats['color'][i], hatch =feats['hatch'][i], edgecolor= 'black')
        plt.bar(oW[i], others[i], width = barWidth, color = feats['color'][i],hatch =feats['hatch'][i], edgecolor= 'black')
    
    else:
        plt.bar(smW[i], smartwatch[i], width = barWidth, color = feats['color'][i], hatch = feats['hatch'][i], edgecolor= 'black', label='2024')
        plt.bar(wbW[i], wristband[i], width = barWidth, color = feats['color'][i], hatch =feats['hatch'][i], edgecolor= 'black')
        plt.bar(oW[i], others[i], width = barWidth, color = feats['color'][i],hatch =feats['hatch'][i], edgecolor= 'black')

        
        
plt.legend()
plt.spines['top'].set_visible(False)
plt.spines['right'].set_visible(False)


plt.set_xticks([])
#plt.set_yticks([0,20,40,60,80] , labelsize = 20)
plt.tick_params(labelsize = 15)
plt.set_ylabel('Shippments in millions', fontsize = 20)
import matplotlib.pyplot as plt
#plt.show()
plt.savefig('wearables.png')
```
