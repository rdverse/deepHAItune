# Test smogn algorithm on artificial data

```python
def dummy_oversample(df):
    datasmogn = smogn.smoter(df, 
                             y='labels',
                             # neighbors to consider   
                             k = 10,                    ## num of neighs for over-sampling (pos int)

                             # For the gaussian noise
                             pert = 0.01,              ## perturbation / noise percentage (pos real)

                             # Use extreme for high over and under sampling
                            samp_method = "extreme",  ## over / under sampling ("balance" or extreme")

                             #Set under sampling to false
                             under_samp = True,        ## under sampling (bool)

                             drop_na_col = False,       ## auto drop columns with nan's (bool)
                             drop_na_row = True,       ## auto drop rows with nan's (bool)
                             replace = False,          ## sampling replacement (bool)

                                ## phi relevance function arguments / inputs
                                rel_thres = 0.75,          ## relevance threshold considered rare (pos real)
                                rel_method = "auto",      ## relevance method ("auto" or "manual")
                                rel_xtrm_type = "low",   ## distribution focus ("high", "low", "both")
                                rel_coef = 2.75,           ## coefficient for box plot (pos real)
                                rel_ctrl_pts_rg = None )   ## input for "manual" rel method  (2d array)
    
    return datasmogn
```

```python
from sklearn.datasets import make_regression
dummy_data, ddl = make_regression(n_samples = 1000,n_features=2, effective_rank =2)
```

```python
plt.scatter(dummy_data[:,0],dummy_data[:,1], c = ddl)
plt.colorbar()
```

```python
dummy_df_data = np.hstack((dummy_data, ddl.reshape(-1,1)))
```

```python
dummy_df = pd.DataFrame(dummy_df_data, columns = ['A','B','labels'])

dummy_over = dummy_oversample(dummy_df)
```
